```python
import pandas as pd
import re

# 1. Define a mapping dictionary from category prefixes to desired main category names
# This dictionary covers the example prefixes provided.
category_mapping = {
    'cs': 'Computer Science',
    'math': 'Mathematics',
    'stat': 'Statistics',
    'astro-ph': 'Astrophysics',
    'hep-ph': 'High Energy Physics - Phenomenology',
    'hep-th': 'High Energy Physics - Theory',
    'quant-ph': 'Quantum Physics',
    'cond-mat': 'Condensed Matter',
    'nucl-exp': 'Nuclear Experiment',
    'nucl-th': 'Nuclear Theory',
    'gr-qc': 'General Relativity and Quantum Cosmology',
    'physics': 'Physics',
    'q-bio': 'Quantitative Biology',
    'q-fin': 'Quantitative Finance',
    'cmp-lg': 'Computation and Language',
    'eess': 'Electrical Engineering and Systems Science',
    'econ': 'Economics',
    'nlin': 'Nonlinear Sciences',
    'chao-dyn': 'Chaos and Nonlinear Dynamics',
    'acc-phys': 'Accelerator Physics'
}

# Define the clean_text function for use in this cell
def clean_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()
        # Remove special characters (keep alphanumeric and spaces)
        text = re.sub(r'[^a-z0-9\s]', '', text)
        # Replace multiple spaces with a single space
        text = re.sub(r'\s+', ' ', text)
        # Strip leading/trailing whitespace
        text = text.strip()
    return text

# 2. Define a function to extract the main category prefix
def get_main_prefix(category_string):
    if not isinstance(category_string, str): # Handle non-string types
        return 'Unknown'
    # Split by space first, to handle cases like 'cs.AI math.ST'
    first_category = category_string.split(' ')[0]
    # Split by '.' and take the first part
    parts = first_category.split('.')

    # Check if the first part is a direct key in category_mapping (e.g., 'cs')
    if parts[0] in category_mapping:
        return parts[0]
    # Check if the entire first_category (e.g., 'astro-ph') is a key
    elif first_category in category_mapping:
        return first_category
    else:
        # Fallback to the first part or 'Unknown' if no clear prefix
        return parts[0] if parts else 'Unknown'


# Re-initialize the df iterator to get the first chunk
if 'data_file_path' in globals() and 'chunk_size' in globals():
    if data_file_path.endswith('.csv'):
        df_reinitialized = pd.read_csv(data_file_path, chunksize=chunk_size)
    elif data_file_path.endswith('.json'):
        df_reinitialized = pd.read_json(data_file_path, lines=True, chunksize=chunk_size)
    else:
        print("Error: Could not re-initialize df. Data file type not recognized or path not set.")
        raise SystemExit("Execution stopped due to unknown data file type.")
else:
    print("Error: 'data_file_path' or 'chunk_size' not found. Please ensure initial data loading cell (EGkRlmfvHbyI) has been executed.")
    raise SystemExit("Execution stopped due to missing data loading variables.")

try:
    # Get only the first chunk
    first_chunk = next(df_reinitialized)
    print("Processing the first chunk...")

    # Select only the 'abstract' and 'categories' columns from the first chunk
    processed_chunk = first_chunk[['abstract', 'categories']].copy()

    # Apply the cleaning function to the 'abstract' column of this chunk
    print("Cleaning 'abstract' column for the current chunk...")
    processed_chunk['abstract'] = processed_chunk['abstract'].apply(clean_text)

    # 3. Apply the function to create a new 'category_prefix' column
    print("Extracting category prefixes from the first chunk...")
    processed_chunk['category_prefix'] = processed_chunk['categories'].apply(get_main_prefix)

    # 4. Map the category prefix to the main category name using the dictionary
    print("Mapping prefixes to main categories for the first chunk...")
    processed_chunk['main_category'] = processed_chunk['category_prefix'].apply(lambda x: category_mapping.get(x, x))

    print("\nFirst chunk with new 'category_prefix' and 'main_category' columns (head):")
    display(processed_chunk.head())

    print("\nCounts of the new 'main_category' in the first chunk (top 10):")
    display(processed_chunk['main_category'].value_counts().head(10))

    print("\nFinished processing the first chunk. No full DataFrame was created.")
except StopIteration:
    print("The dataframe iterator 'df_reinitialized' was empty or already exhausted. Please ensure the data loading cell (EGkRlmfvHbyI) was run and produced data.")
except Exception as e:
    print(f"An error occurred during processing the first chunk: {e}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from sklearn.metrics import accuracy_score

# Ensure X_train, X_test, y_train, y_test, model, criterion, and optimizer are available
if 'X_train' not in locals() or 'y_train' not in locals() or 'model' not in locals() or 'criterion' not in locals() or 'optimizer' not in locals():
    print("Error: Model components (X_train, y_train, model, criterion, optimizer) not defined. Please run previous cells.")
else:
    print("Starting model training...")

    # Convert sparse TF-IDF matrices to dense tensors
    # PyTorch's nn.Linear expects dense tensors
    X_train_dense = X_train.toarray() if hasattr(X_train, 'toarray') else X_train
    X_test_dense = X_test.toarray() if hasattr(X_test, 'toarray') else X_test

    # Convert numpy arrays to PyTorch tensors
    X_train_tensor = torch.tensor(X_train_dense, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train, dtype=torch.long) # Use torch.long for CrossEntropyLoss labels
    X_test_tensor = torch.tensor(X_test_dense, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test, dtype=torch.long)

    # Create TensorDatasets
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    # Create DataLoaders
    batch_size = 64
    train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)

    print(f"DataLoaders created with batch size: {batch_size}")
    print(f"Number of training batches: {len(train_loader)}")
    print(f"Number of testing batches: {len(test_loader)}")

    # Check for GPU availability and move model/data to GPU if available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    print(f"Using device: {device}")

    # Training loop
    num_epochs = 10
    for epoch in range(num_epochs):
        model.train() # Set model to training mode
        running_loss = 0.0
        correct_predictions = 0
        total_predictions = 0

        for i, (inputs, labels) in enumerate(train_loader):
            inputs, labels = inputs.to(device), labels.to(device)

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Backward pass and optimize
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            # Calculate training accuracy
            _, predicted = torch.max(outputs.data, 1)
            total_predictions += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()

            if (i + 1) % 100 == 0: # Print every 100 batches
                print(f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{len(train_loader)}], Loss: {loss.item():.4f}, Acc: {(correct_predictions/total_predictions)*100:.2f}%')

        epoch_loss = running_loss / len(train_loader)
        epoch_accuracy = (correct_predictions / total_predictions) * 100
        print(f'Epoch [{epoch+1}/{num_epochs}] finished. Average Loss: {epoch_loss:.4f}, Training Accuracy: {epoch_accuracy:.2f}%')

        # Evaluation on test set after each epoch
        model.eval() # Set model to evaluation mode
        with torch.no_grad(): # Disable gradient calculation for evaluation
            test_correct = 0
            test_total = 0
            all_predictions = []
            all_labels = []
            for inputs, labels in test_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs.data, 1)
                test_total += labels.size(0)
                test_correct += (predicted == labels).sum().item()
                all_predictions.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

            test_accuracy = (test_correct / test_total) * 100
            print(f'Epoch [{epoch+1}/{num_epochs}] Test Accuracy: {test_accuracy:.2f}%')

    print("\nTraining complete!")

    # Final evaluation (optional, but good for summary)
    print("\nPerforming final evaluation on the test set...")
    model.eval()
    with torch.no_grad():
        final_test_correct = 0
        final_test_total = 0
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            final_test_total += labels.size(0)
            final_test_correct += (predicted == labels).sum().item()
        final_test_accuracy = (final_test_correct / final_test_total) * 100
        print(f'Final Test Accuracy: {final_test_accuracy:.2f}%')

    # Save the trained model (optional)
    # torch.save(model.state_dict(), 'text_classification_model.pth')
    # print("Model saved to text_classification_model.pth")
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from scipy.sparse import vstack

# --- Re-define functions and mapping for self-containment ---
category_mapping = {
    'cs': 'Computer Science',
    'math': 'Mathematics',
    'stat': 'Statistics',
    'astro-ph': 'Astrophysics',
    'hep-ph': 'High Energy Physics - Phenomenology',
    'hep-th': 'High Energy Physics - Theory',
    'quant-ph': 'Quantum Physics',
    'cond-mat': 'Condensed Matter',
    'nucl-exp': 'Nuclear Experiment',
    'nucl-th': 'Nuclear Theory',
    'gr-qc': 'General Relativity and Quantum Cosmology',
    'physics': 'Physics',
    'q-bio': 'Quantitative Biology',
    'q-fin': 'Quantitative Finance',
    'cmp-lg': 'Computation and Language',
    'eess': 'Electrical Engineering and Systems Science',
    'econ': 'Economics',
    'nlin': 'Nonlinear Sciences',
    'chao-dyn': 'Chaos and Nonlinear Dynamics',
    'acc-phys': 'Accelerator Physics'
}

def clean_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[^a-z0-9\s]', '', text)
        text = re.sub(r'\s+', ' ', text)
        text = text.strip()
    return text

def get_main_prefix(category_string):
    if not isinstance(category_string, str):
        return 'Unknown'
    first_category = category_string.split(' ')[0]
    parts = first_category.split('.')
    if parts[0] in category_mapping:
        return parts[0]
    elif first_category in category_mapping:
        return first_category
    else:
        return parts[0] if parts else 'Unknown'

# --- Re-initialize df iterator for full dataset processing ---
if 'data_file_path' in globals() and 'chunk_size' in globals():
    if data_file_path.endswith('.csv'):
        df_full_iterator = pd.read_csv(data_file_path, chunksize=chunk_size)
    elif data_file_path.endswith('.json'):
        df_full_iterator = pd.read_json(data_file_path, lines=True, chunksize=chunk_size)
    else:
        print("Error: Could not re-initialize df. Data file type not recognized or path not set.")
        raise SystemExit("Execution stopped due to unknown data file type.")
else:
    print("Error: 'data_file_path' or 'chunk_size' not found. Please ensure initial data loading cell (EGkRlmfvHbyI) has been executed.")
    raise SystemExit("Execution stopped due to missing data loading variables.")

print("Processing all chunks of the dataset...")

all_tfidf_matrices = []
all_main_categories = []

# Initialize TF-IDF Vectorizer once for the entire dataset
# This will fit on all 'abstract' data as it comes in chunks
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

# Collect abstracts for fitting TF-IDF first, then transform
# This is more memory-efficient than trying to fit_transform sparse matrices directly
# or storing all abstracts in memory before fitting. We'll fit_transform on each chunk later.
# However, for a proper TF-IDF, it's best to fit on the entire corpus.
# Let's collect all abstracts first to fit the vectorizer properly.

# To fit TF-IDF on the entire corpus, we need to gather all abstracts first.
# This might be memory intensive for very large datasets if all abstracts are stored.
# Given chunking, we'll try a common approach: fit on the first N chunks or a sample.
# A more robust approach for extremely large datasets would be Incremental TF-IDF or fitting on a representative sample.
# For now, let's assume fit_transform on individual chunks and vstack is acceptable.
# However, if tfidf_vectorizer was already fitted on a single chunk as in the notebook, this is not ideal for the whole dataset.
# The best practice is to fit TF-IDF on the full training data before transforming.
# Re-fitting it here on the first chunk just like before would be wrong.
# A better way is to iterate twice: once to fit the vectorizer, once to transform.

# Let's re-run the data preparation cells first to ensure tfidf_vectorizer is ready.
# The previous cell gBc2Z7HAzrBH already fitted tfidf_vectorizer on the first chunk.
# To truly use the whole data, we need to fit the TF-IDF on the whole data as well.
# This means we need to re-create the tfidf_vectorizer using the full dataset. This will require two passes or more memory.

# Strategy: Collect all abstracts, fit TF-IDF, then re-iterate to transform.
# This can be memory intensive. Alternative: Fit TF-IDF incrementally (not directly supported by scikit-learn's TfidfVectorizer).
# A pragmatic approach for chunks: use partial_fit or fit on a larger sample, then transform all.
# Given the current notebook structure, the tfidf_vectorizer is fitted on processed_chunk (first chunk).
# To be consistent with the user's request, we should fit TF-IDF on all abstracts.

print("Collecting all abstracts to fit TF-IDF vectorizer...")
all_abstracts = []
# Re-initialize the iterator again to get all chunks
if data_file_path.endswith('.csv'):
    df_abstract_collector = pd.read_csv(data_file_path, chunksize=chunk_size)
elif data_file_path.endswith('.json'):
    df_abstract_collector = pd.read_json(data_file_path, lines=True, chunksize=chunk_size)

for i, chunk in enumerate(df_abstract_collector):
    # Ensure 'abstract' column exists
    if 'abstract' in chunk.columns:
        cleaned_abstracts = chunk['abstract'].apply(clean_text)
        all_abstracts.extend(cleaned_abstracts.tolist())
    print(f"Collected abstracts from chunk {i+1}...")

print(f"Fitting TF-IDF vectorizer on {len(all_abstracts)} abstracts...")
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
tfidf_vectorizer.fit(all_abstracts)
del all_abstracts # Free up memory
print("TF-IDF vectorizer fitted on the complete dataset.")

# Re-initialize the iterator one last time for feature extraction and label encoding
if data_file_path.endswith('.csv'):
    df_full_iterator = pd.read_csv(data_file_path, chunksize=chunk_size)
elif data_file_path.endswith('.json'):
    df_full_iterator = pd.read_json(data_file_path, lines=True, chunksize=chunk_size)

for i, chunk in enumerate(df_full_iterator):
    # Apply preprocessing to the current chunk
    processed_chunk = chunk[['abstract', 'categories']].copy()
    processed_chunk['abstract'] = processed_chunk['abstract'].apply(clean_text)
    processed_chunk['category_prefix'] = processed_chunk['categories'].apply(get_main_prefix)
    processed_chunk['main_category'] = processed_chunk['category_prefix'].apply(lambda x: category_mapping.get(x, x))

    # Transform abstracts using the fitted TF-IDF vectorizer
    chunk_tfidf_matrix = tfidf_vectorizer.transform(processed_chunk['abstract'])
    all_tfidf_matrices.append(chunk_tfidf_matrix)
    all_main_categories.extend(processed_chunk['main_category'].tolist())

    print(f"Processed chunk {i+1} and extracted features/labels.")

print("Concatenating all TF-IDF matrices and labels...")
X_full = vstack(all_tfidf_matrices) # Combine all sparse matrices vertically
y_full = pd.Series(all_main_categories) # Create a pandas Series for labels

# Encode labels using LabelEncoder
label_encoder = LabelEncoder()
y_encoded_full = label_encoder.fit_transform(y_full)

print(f"Full TF-IDF matrix shape: {X_full.shape}")
print(f"Full encoded labels shape: {y_encoded_full.shape}")
print(f"Number of unique categories: {len(label_encoder.classes_)}")

# Split the full dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_full, y_encoded_full, test_size=0.2, random_state=42, stratify=y_encoded_full
)

print("\nData split into training and testing sets for the full dataset:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape:  {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape:  {y_test.shape}")

print("\nFull dataset preparation complete. You can now re-run the model training (cell 4A75cwbw5zfL) and evaluation (cell mkMi36rt5zbl) cells.")
```

```python
# ============================================================
# PART 2 — MODEL DEFINITION + TRAINING LOOP
# Connects directly to your Part 1 variables:
# X_train, X_test, y_train, y_test, label_encoder
# ============================================================

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, classification_report

# ---------------------------------------------------------------
# CHECK PART 1 VARIABLES EXIST
# Same check as your existing code
# ---------------------------------------------------------------
if 'X_train' not in globals() or 'y_train' not in globals() or 'label_encoder' not in globals():
    print("Error: X_train, y_train or label_encoder not found. Please run Part 1 first.")
    raise SystemExit("Stopped.")

print("All Part 1 variables found. Starting Part 2...")
print(f"X_train shape : {X_train.shape}")
print(f"X_test shape  : {X_test.shape}")
print(f"y_train shape : {y_train.shape}")
print(f"y_test shape  : {y_test.shape}")
print(f"Categories    : {len(label_encoder.classes_)}")


# ---------------------------------------------------------------
# DEVICE SETUP
# ---------------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\nUsing device: {device}")


# ---------------------------------------------------------------
# SPARSE DATASET CLASS
# Converts TF-IDF sparse rows to dense tensors on the fly
# Avoids loading the full dense matrix into RAM at once
# ---------------------------------------------------------------
class SparseDataset(Dataset):
    def __init__(self, X_sparse, y):
        self.X_sparse = X_sparse
        self.y        = np.asarray(y)

    def __len__(self):
        return self.X_sparse.shape[0]

    def __getitem__(self, idx):
        # Convert one sparse row to dense float tensor
        x     = self.X_sparse[idx].toarray().astype(np.float32)
        x     = torch.tensor(x).squeeze(0)
        label = torch.tensor(self.y[idx], dtype=torch.long)
        return x, label


# ---------------------------------------------------------------
# CREATE DATASETS AND DATALOADERS
# Same batch size as your existing code
# ---------------------------------------------------------------
train_dataset = SparseDataset(X_train, y_train)
test_dataset  = SparseDataset(X_test,  y_test)

batch_size    = 64
train_loader  = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader   = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

print(f"\nDataLoaders created with batch size : {batch_size}")
print(f"Training batches                    : {len(train_loader)}")
print(f"Testing batches                     : {len(test_loader)}")


# ---------------------------------------------------------------
# MODEL DEFINITION
# Same architecture — takes TF-IDF features as input
# ---------------------------------------------------------------
class SimpleNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(SimpleNN, self).__init__()

        self.network = nn.Sequential(
            nn.Linear(input_dim, 512),  # Input → Hidden Layer 1
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(512, 256),        # Hidden Layer 1 → Hidden Layer 2
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(256, 128),        # Hidden Layer 2 → Hidden Layer 3
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, num_classes) # Hidden Layer 3 → Output
        )

    def forward(self, x):
        return self.network(x)


# ---------------------------------------------------------------
# INSTANTIATE MODEL, LOSS, OPTIMIZER
# input_dim comes directly from your X_train shape
# num_classes comes directly from your label_encoder
# ---------------------------------------------------------------
input_dim   = X_train.shape[1]              # 5000 from your TF-IDF max_features
num_classes = len(label_encoder.classes_)   # 18-20 from your label encoder

model     = SimpleNN(input_dim, num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print(f"\nModel created")
print(f"Input dimension  : {input_dim}")
print(f"Output classes   : {num_classes}")
print(model)


# ---------------------------------------------------------------
# TRAINING LOOP
# Same structure as your existing code
# ---------------------------------------------------------------
num_epochs = 10
print(f"\nStarting training for {num_epochs} epochs...")

for epoch in range(num_epochs):
    model.train()
    running_loss        = 0.0
    correct_predictions = 0
    total_predictions   = 0

    for i, (inputs, labels) in enumerate(train_loader):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss    = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        _, predicted       = torch.max(outputs.data, 1)
        total_predictions  += labels.size(0)
        correct_predictions += (predicted == labels).sum().item()

        # Print every 100 batches — same as your existing code
        if (i + 1) % 100 == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}] "
                  f"Step [{i+1}/{len(train_loader)}] "
                  f"Loss: {loss.item():.4f} "
                  f"Acc: {correct_predictions/total_predictions*100:.2f}%")

    epoch_loss     = running_loss / len(train_loader)
    epoch_accuracy = correct_predictions / total_predictions * 100
    print(f"\nEpoch [{epoch+1}/{num_epochs}] finished "
          f"— Loss: {epoch_loss:.4f} "
          f"— Train Acc: {epoch_accuracy:.2f}%")

    # ── Evaluate after every epoch — same as your existing code ──
    model.eval()
    test_correct     = 0
    test_total       = 0
    all_predictions  = []
    all_labels_list  = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs        = model(inputs)
            _, predicted   = torch.max(outputs.data, 1)
            test_total    += labels.size(0)
            test_correct  += (predicted == labels).sum().item()
            all_predictions.extend(predicted.cpu().numpy())
            all_labels_list.extend(labels.cpu().numpy())

    test_accuracy = test_correct / test_total * 100
    print(f"Epoch [{epoch+1}/{num_epochs}] "
          f"Test Accuracy: {test_accuracy:.2f}%\n")


# ---------------------------------------------------------------
# FINAL EVALUATION
# Same as your existing final evaluation block
# ---------------------------------------------------------------
print("\nPerforming final evaluation on test set...")
model.eval()
final_correct = 0
final_total   = 0
final_preds   = []
final_labels  = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs        = model(inputs)
        _, predicted   = torch.max(outputs.data, 1)
        final_total   += labels.size(0)
        final_correct += (predicted == labels).sum().item()
        final_preds.extend(predicted.cpu().numpy())
        final_labels.extend(labels.cpu().numpy())

final_accuracy = final_correct / final_total * 100
print(f"Final Test Accuracy : {final_accuracy:.2f}%")

# F1 score across all categories
f1 = f1_score(final_labels, final_preds, average='weighted')
print(f"Weighted F1 Score   : {f1:.4f}")

# Full breakdown per category
print("\nClassification Report:")
print(classification_report(
    final_labels,
    final_preds,
    target_names=label_encoder.classes_
))


# ---------------------------------------------------------------
# SAVE MODEL TO GOOGLE DRIVE
# ---------------------------------------------------------------
torch.save(model.state_dict(), '/content/drive/MyDrive/tfidf_nn_model.pth')
print("\nModel saved to Google Drive at: /content/drive/MyDrive/tfidf_nn_model.pth")
```

```python
import torch
import pickle
import torch.nn as nn

# -----------------------------
# Define model (same as training)
# -----------------------------
class SimpleNN(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(SimpleNN, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Dropout(0.5),

            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.network(x)


# -----------------------------
# LOAD EVERYTHING (runs once)
# -----------------------------
def load_model_and_vectorizer():
    # Load vectorizer
    with open('tfidf_vectorizer.pkl', 'rb') as f:
        vectorizer = pickle.load(f)

    # Load model
    checkpoint = torch.load('tfidf_nn_model.pth', map_location='cpu')

    model = SimpleNN(checkpoint['input_dim'], checkpoint['num_classes'])
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()

    classes = checkpoint['classes']

    return model, vectorizer, classes


# -----------------------------
# PREDICTION FUNCTION
# -----------------------------
def predict_abstract(text):
    model, vectorizer, classes = load_model_and_vectorizer()

    # Convert text → TF-IDF
    X = vectorizer.transform([text])
    X = X.toarray().astype('float32')

    X_tensor = torch.tensor(X)

    with torch.no_grad():
        outputs = model(X_tensor)
        _, predicted = torch.max(outputs, 1)

    return classes[predicted.item()]
```
